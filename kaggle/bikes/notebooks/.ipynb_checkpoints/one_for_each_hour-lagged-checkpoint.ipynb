```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestRegressor

def get_features(data):
    return data[['season',
                 'holiday',
                 'workingday',
                 'weather',
                 'temp',
                 'atemp',
                 'humidity',
                 'windspeed',
                 'prev',
                 'next']]

def update_df(data):
    dt = pd.DatetimeIndex(data['datetime'])
    data['hour'] = dt.hour
    data['month'] = dt.month
    data['weekday'] = dt.weekday

def erro(actual, predicted):
    num = len(actual)
    err = sum([(np.log(predicted[i]+1)-np.log(actual[i]+1))**2 
               for i in range(num)])
    err /= num
    err = np.sqrt(err)
    return err
```

```python
def pred(model, data):
    predictions = np.array(model.predict(np.array(get_features(data))))
    dt = data['datetime']
    predictions = pd.Series(predictions, index=dt.index)
    res = pd.concat([dt, predictions], axis=1)
    res.columns=['datetime', 'count']
    return res

train = pd.read_csv('../train.csv', parse_dates=['datetime'])
dt = pd.DatetimeIndex(train['datetime']).day
intest = train[dt>15]
train = train[dt<=15]
#test = pd.read_csv('../test.csv', parse_dates=['datetime'])
update_df(train)
update_df(intest)
#update_df(test)

inresults = pd.DataFrame(columns=['datetime', 'count'])
#results = pd.DataFrame(columns=['datetime', 'count'])
for hour , intest_subset in intest.groupby('hour'):
    train_subset = train[train['hour']==hour]
    #intest_subset = intest[intest['hour']==hour]
    model = RandomForestRegressor(n_estimators=100)
    model.fit(np.array(get_features(train_subset)), np.array(train_subset['count']))
    #results = pd.concat([results, pred(model, test_subset)])
    inresults = pd.concat([inresults, pred(model, intest_subset)])
    
inresults['count'] = inresults['count'].astype('int64')
mrg = pd.merge(inresults, intest, on='datetime')
print erro(mrg['count_x'].values, mrg['count_y'].values)
#results['count'] = results['count'].astype('int64')
#results = results.sort('datetime')
#results.to_csv('../submissions/sixthSubmission.csv', index=False)
```

```python
from pandas.tseries.offsets import DateOffset

train = pd.read_csv('../train.csv', parse_dates=['datetime'])
test = pd.read_csv('../test.csv', parse_dates=['datetime'])  
update_df(train)
update_df(test)

for name in ['prev', 'next']:
    train[name] = pd.Series(None, index=train.index)
    
def add_days(index, frame):
    current_time = frame.ix[idx, 'datetime']
    prev_day = current_time + DateOffset(days=-1)
    next_day = current_time + DateOffset(days=1)
    lag_days = [(prev_day, 'prev'), 
                (next_day, 'next')]
    for date, name in lag_days:
        bl = (frame['datetime']==date)
        if bl.any():
            frame.ix[idx,name] = frame[bl]['count'].values[0]
        else:
            frame.ix[idx, name] = frame.ix[idx,'count']
 
for idx in train.index:
    add_days(idx, train)
print 'days added to train'
#aggr contains all dates in train and test
aggr = train.copy()
 
aggr = pd.concat([aggr, test])
print 'test added to aggr'
for idx in aggr.index:
    if aggr.ix[idx, 'prev'] == None or aggr.ix[idx, 'next'] == 'None':
        print '.', 
        add_days(idx, aggr)

for hour , test_subset in test.groupby('hour'):
    train_subset = train[train['hour']==hour]
    model = RandomForestRegressor(n_estimators=100)
    model.fit(np.array(get_features(train_subset)), np.array(train_subset['count']))
    for idx in test_subset.index:
        aggr_idx = test[test['datetime'==test_subset.ix[idx, 'datetime']]].index[0]
        prediction = model.predict(np.array(get_features(aggr.ix[aggr_idx])))
        print prediction
        exit
        aggr_idx.ix[aggr_idx, 'count'] = prediction
        print aggr_idx
        
```
